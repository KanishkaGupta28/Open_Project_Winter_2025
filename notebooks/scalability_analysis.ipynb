```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load CSV (same folder as notebook)
data = pd.read_csv("scalability_results.csv")

print("Scalability Results:")
display(data)

# Create results folder if it does not exist
os.makedirs("results", exist_ok=True)

# Plot 1: Fidelity
plt.figure(figsize=(6,4))
plt.errorbar(
    data["n_qubits"],
    data["mean_fidelity"],
    yerr=data["fidelity_std"],
    marker='o',
    capsize=5
)

plt.xlabel("Number of Qubits")
plt.ylabel("Mean Fidelity")
plt.title("Fidelity vs Number of Qubits")
plt.grid(True)

plt.savefig("results/fidelity_plot.png", dpi=300)
plt.show()


# Plot 2: Runtime
plt.figure(figsize=(6,4))
plt.plot(
    data["n_qubits"],
    data["mean_runtime"],
    marker='o'
)

plt.xlabel("Number of Qubits")
plt.ylabel("Mean Runtime (seconds)")
plt.title("Runtime vs Number of Qubits")
plt.grid(True)

plt.savefig("results/runtime_plot.png", dpi=300)
plt.show()

print("Plots saved inside 'results' folder.")

```
